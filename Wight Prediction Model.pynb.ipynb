```python
import pandas as pd #importing pandas library and assigning it to pd
import numpy as np
import matplotlib.pyplot as plt
import warnings as wr
wr.filterwarnings('ignore') #ignoring all the unnecessary warning message

df = pd.read_csv("Height-Weight Data.csv", sep = ',') #rading the csv file into df using the pandas library
df.head()
```

```python
plt.scatter (df.Height, df.Weight, color = 'red', marker = '+') #defining axis, marker and the color to make the scatter plot
#x always comes in the first place and y goes in the second place (#common sense x, y) here to define the axis we use df.(the variable)
plt.xlabel('Height')
plt.ylabel('Weight')
plt.title('Height-Weight Comparison')
```

```python
from sklearn.linear_model import LinearRegression #importing linearregression from sklearn
reg = LinearRegression()
from sklearn.model_selection import train_test_split as tts #to import splitting data model from library

x = df[['Height']]
y = df[['Weight']]

xtrain, xtest, ytrain, ytest = tts(x,y, test_size = 0.33, random_state = 1) #splitting data on test and trainging basis. Here, 33 % of data gonna be
#in test size while rest of the data will be counted as the training
xtrain.head() #train value of x
```

```python
reg.fit(xtrain, ytrain) #training model
```

```python
m = reg.coef_
c = reg.intercept_

m
```

```python
c
```

```python
predicted = (m*191.563495) + c   #cheking the value of y when the value of x is 191.563495
predicted
```

```python
plt.scatter (df.Height, df.Weight, color = 'green', marker = '+') #defining axis, marker and the color to make the scatter plot
#x always comes in the first place and y goes in the second place (#common sense x, y) here to define the axis we use df.(the variable)
plt.plot(xtrain, reg.predict(xtrain), color ='red') #predicting the result based on the xtrain, here xtrain is the independent variable where reg.predict is the y variable
plt.xlabel('Height')
plt.ylabel('Weight')
plt.title('Height-Weight Comparison')
```

```python
#importing MAE and MSE
from sklearn.metrics import mean_squared_error as mse
from sklearn.metrics import mean_absolute_error as mae

test_mse = mse(ytest, reg.predict(xtest))
test_mse
```

```python
test_mae = mae(ytest, reg.predict(xtest))
test_mae
```
